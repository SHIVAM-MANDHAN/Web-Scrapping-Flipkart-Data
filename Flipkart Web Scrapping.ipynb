```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url= "https://www.flipkart.com/search?q=smart+tv&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&as-pos=1&as-type=RECENT&suggestionId=smart+tv%7CTelevisions&requestId=6555d214-ff1c-4beb-aa1a-eaec991ce156&as-searchtext=smart%20tv"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
model_code = soup.find_all("div", class_ = "_4rR01T")

model = []
for i in range(len(model_code)):
    temp_model = model_code[i].text
    model.append(temp_model)

model[:5]
```

```python
model_name = []

for i in model:
    name = i.split("(")
    model_name.append(name[0])
    
product = []
size = []

for i in model_name:
    product.append(i.split()[:-2])
    size.append(i.split()[-2:])
    
product1 = []
for i in product:
    product1.append(i[0])
    
size1 = []
for i in size:
    size1.append(i[0])
```

```python
price_code = soup.find_all("div", class_ = "_30jeq3 _1_WHN1")

price = []
for i in range(len(price_code)):
    temp_price = price_code[i].text
    price.append(temp_price)

price[:5]
```

```python
rating_code = soup.find_all("div", class_ = "_3LWZlK")

rating = []
for i in range(len(rating_code)):
    temp_rating = rating_code[i].text
    rating.append(temp_rating)

```

```python
total_rating_code = soup.find_all("span", class_ = "_2_R_DZ")

total_rating = []
for i in range(len(total_rating_code)):
    temp_total_rating = total_rating_code[i].text
    total_rating.append(temp_total_rating)

total_rating[:5]
```

```python
rating_numbers = []
review_numbers = []

for i in total_rating:
    temp = i.split()
    rating_numbers.append(temp[0])
    review_numbers.append(temp[3])
    
rating_numbers[:5], review_numbers[:5]
```

```python
description_code = soup.find_all("li", class_ = "rgWa7D")

description = []
for i in range(len(description_code)):
    temp_description = description_code[i].text
    description.append(temp_description)

description[:6]
```

```python
os = []

for i in range(0, len(description), 3):
    temp_os = description[i]
    os.append(temp_os)
    
screen = []

for i in range(1, len(description), 3):
    temp_screen = description[i]
    screen.append(temp_screen)
    
warranty = []

for i in range(2, len(description), 3):
    temp_warrany = description[i]
    warranty.append(temp_warrany)
```

```python
df = pd.DataFrame()
```

```python
df["Product"] = product1
df["Size (cm)"] = size1
df["Price"] = price
# df["Rating"] = rating
# df["Total_ratings"] = rating_numbers
# df["Total_reviews"] = review_numbers
df["Operating System"] = os
df["Screen"] = screen
df["Warranty"] = warranty
df
```
